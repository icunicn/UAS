### 2.3 Menggabungkan semua data sheet menjadi satu

```
xl = pd.ExcelFile(file_path)
cols = ['No', 'Nama', 's.akhir', 's.awal',
        'Pakai', '/meter', 'Total', 'Admin', 'Total akhir']
BULAN_ORDER = ['JANUARI','FEBRUARI','MARET','APRIL','MEI','JUNI',
                'JULI','AGUSTUS','SEPTEMBER']

all_dfs = []
for sheet in xl.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet, header=3)
    df = df.dropna(axis=1, how='all')
    df = df[cols].copy()
    df = df.dropna(subset=['No'])
    df = df[pd.to_numeric(df['No'], errors='coerce').notna()]
    df['Bulan'] = sheet
    all_dfs.append(df)

combined = pd.concat(all_dfs, ignore_index=True)
combined['No'] = combined['No'].astype(int)
combined = combined[['Bulan'] + cols]


print(f"Total baris: {len(combined)}")
display(combined.head(10))
```

### 2.5 Data Cleaning
Memeriksa apakah ada data yang tidak normal

```
combined = pd.concat(all_dfs, ignore_index=True)
combined['No'] = combined['No'].astype(int)
combined['Pakai'] = pd.to_numeric(combined['Pakai'], errors='coerce')
combined['Total akhir'] = pd.to_numeric(combined['Total akhir'], errors='coerce')
combined = combined[['Bulan'] + cols]

df = combined[combined['Pakai'] >= 0 ].copy()

combined['Segmen'] = combined['/meter'].map({500: 'Rumah Tangga Biasa', 1000: 'Rumah Tangga Produksi'})

print(f'Total baris (raw)  : {len(combined):,}')
print(f'Total baris (bersih): {len(df):,}')
print(f'Nilai negatif      : {(combined["Pakai"] < 0).sum()} baris (potensi data null)')
print(combined['/meter'].value_counts())
df.head()
```

## Analisis Data Lanjutan

```
def iqr_outliers(x):
    q1, q3 = x.quantile(0.25), x.quantile(0.75)
    iqr = q3 - q1
    return ((x < q1 - 1.5*iqr) | (x > q3 + 1.5*iqr)).sum()

stats = df.groupby('Bulan')['Pakai'].agg(
    n='count',
    mean='mean',
    median='median',
    std='std',
    min='min',
    max='max',
    q25=lambda x: x.quantile(0.25),
    q75=lambda x: x.quantile(0.75),
).round(1)

stats['outliers']  = df.groupby('Bulan')['Pakai'].apply(iqr_outliers).values
stats['zero_pct']  = (df.groupby('Bulan')['Pakai'].apply(lambda x: (x==0).sum())
                       / stats['n'] * 100).round(1).values
stats = stats.reindex(BULAN_ORDER)
stats.index.name = 'Bulan'

print('Statistik Deskriptif Pemakaian Air per Bulan (m³):')
print(stats.to_string())
```

```
fig = go.Figure()

fig.add_trace(go.Bar(
    x=BULAN_ORDER, y=stats['mean'],
    name='Rata-rata', marker_color='#378ADD', opacity=0.85
))
fig.add_trace(go.Scatter(
    x=BULAN_ORDER, y=stats['median'],
    name='Median', mode='lines+markers',
    line=dict(color='#D85A30', width=2.5),
    marker=dict(size=8)
))

fig.update_layout(
    title='Rata-rata vs Median Pemakaian Air per Bulan',
    xaxis_title='Bulan', yaxis_title='Pemakaian (m³)',
    legend=dict(orientation='h', yanchor='bottom', y=1.02),
    plot_bgcolor='white', height=420,
    xaxis=dict(tickangle=-30)
)
fig.update_xaxes(showgrid=False)
fig.update_yaxes(gridcolor='#f0f0f0')
fig.show()
```

```
print('=== TOP 3 PEMAKAIAN TERTINGGI (OUTLIER) PER BULAN ===')
for b in BULAN_ORDER:
    sub  = df[df['Bulan'] == b]
    q1, q3 = sub['Pakai'].quantile(0.25), sub['Pakai'].quantile(0.75)
    iqr  = q3 - q1
    batas = q3 + 1.5 * iqr
    outliers = sub[sub['Pakai'] > batas].nlargest(3, 'Pakai')[['Nama','Pakai']]
    print(f'\n{b} (batas atas: {batas:.0f} m³):')
    for _, r in outliers.iterrows():
        print(f'  {r["Nama"]:<30} {r["Pakai"]:>6.0f} m³')
```

```

print('=== Deteksi Outlier per Segmen ===')
for seg in ['Rumah Tangga Biasa', 'Rumah Tangga Produksi']:
    sub  = df[df['Segmen']==seg]['Pakai']
    q1, q3 = sub.quantile(0.25), sub.quantile(0.75)
    batas = q3 + 1.5*(q3-q1)
    out   = df[(df['Segmen']==seg) & (df['Pakai'] > batas)]
    print(f'\n{seg}')
    print(f'  Q1={q1:.0f}, Q3={q3:.0f}, Batas atas={batas:.0f} m³')
    print(f'  Outlier: {len(out)} dari {len(df[df["Segmen"]==seg])} rekening ({len(out)/len(df[df["Segmen"]==seg])*100:.1f}%)')
    print(f'  Top 5:')
    top5 = out.nlargest(5,'Pakai')[['Bulan','Nama','Pakai','Total akhir']]
    print(top5.to_string(index=False))
```

# Bussiness Understanding



```
# ── 4.1 Revenue Mix: Biasa vs Produksi ──────────────────────────────────────
rev = df.groupby('Segmen')['Total akhir'].sum()
n_pel = df.groupby('Segmen')['Nama'].nunique()
total_rev = rev.sum()

print('=== Revenue Mix ===')
for seg in rev.index:
    arpu = rev[seg] / n_pel[seg]
    print(f'{seg}')
    print(f'  Pelanggan  : {n_pel[seg]}')
    print(f'  Revenue    : Rp {rev[seg]:>15,.0f}  ({rev[seg]/total_rev*100:.1f}%)')
    print(f'  ARPU/tahun : Rp {arpu:>15,.0f}')
    print(f'  ARPU/bulan : Rp {arpu/10:>15,.0f}')
    print()
print(f'Total Revenue Keseluruhan: Rp {total_rev:,.0f}')

# Pie chart
fig = go.Figure(go.Pie(
    labels=rev.index.tolist(),
    values=rev.values,
    hole=0.45,
    marker_colors=[WARNA_BIASA, WARNA_PRODUKSI],
    textinfo='label+percent+value',
    texttemplate='%{label}<br>Rp %{value:,.0f}<br>(%{percent})'
))
fig.update_layout(
    title='Komposisi Revenue: Biasa vs Produksi',
    height=420, showlegend=False
)
fig.show()
```

```
# ── 4.2 Revenue Bulanan per Segmen ──────────────────────────────────────────
rev_bulan = df.groupby(['Bulan','Segmen'], observed=True)['Total akhir'].sum().reset_index()
rev_bulan['Bulan'] = rev_bulan['Bulan'].astype(str)

fig = go.Figure()
for seg, color in [('Rumah Tangga Biasa', WARNA_BIASA),
                    ('Rumah Tangga Produksi', WARNA_PRODUKSI)]:
    sub = rev_bulan[rev_bulan['Segmen']==seg]
    fig.add_trace(go.Bar(
        x=sub['Bulan'], y=sub['Total akhir'],
        name=seg, marker_color=color
    ))

fig.update_layout(
    title='Revenue Bulanan per Segmen Tarif',
    barmode='stack', yaxis_title='Revenue (Rp)',
    plot_bgcolor='white', height=430,
    xaxis=dict(tickangle=-30),
    legend=dict(orientation='h', y=1.08),
    yaxis_tickformat=',.0f'
)
fig.update_yaxes(gridcolor='#f0f0f0')
fig.show()
```

```
# ── 4.4 Segmentasi Pelanggan: Ringan / Sedang / Berat ────────────────────────
# Hitung rata-rata pemakaian bulanan tiap pelanggan (per segmen)
profil = df.groupby(['Nama','Segmen']).agg(
    avg_pakai=('Pakai','mean'),
    total_pakai=('Pakai','sum'),
    total_tagihan=('Total akhir','sum'),
    bulan_aktif=('Bulan','count')
).reset_index().round(1)

# Kategori konsumsi berdasarkan rata-rata per bulan
def kategori(avg, segmen):
    if segmen == 'Rumah Tangga Produksi':
        if avg <= 50:  return 'Produksi Ringan'
        if avg <= 150: return 'Produksi Sedang'
        return 'Produksi Berat'
    else:
        if avg <= 0:   return 'Tidak Pakai'
        if avg <= 15:  return 'Hemat'
        if avg <= 40:  return 'Normal'
        if avg <= 80:  return 'Tinggi'
        return 'Sangat Tinggi'

profil['kategori'] = profil.apply(lambda r: kategori(r['avg_pakai'], r['Segmen']), axis=1)

print('=== Segmentasi Pelanggan berdasarkan Rata-rata Pemakaian ===')
print(profil['kategori'].value_counts().to_string())

# Visualisasi
cat_order_biasa = ['Tidak Pakai','Hemat','Normal','Tinggi','Sangat Tinggi']
cat_order_prod  = ['Produksi Ringan','Produksi Sedang','Produksi Berat']

for seg, cat_order, color in [
    ('Rumah Tangga Biasa', cat_order_biasa, WARNA_BIASA),
    ('Rumah Tangga Produksi', cat_order_prod, WARNA_PRODUKSI)
]:
    sub  = profil[profil['Segmen']==seg]['kategori'].value_counts().reindex(cat_order, fill_value=0)
    fig  = go.Figure(go.Bar(
        x=sub.index, y=sub.values,
        text=sub.values, textposition='outside',
        marker_color=color
    ))
    fig.update_layout(
        title=f'Segmentasi Pelanggan — {seg}',
        yaxis_title='Jumlah Pelanggan',
        plot_bgcolor='white', height=380
    )
    fig.update_yaxes(gridcolor='#f0f0f0')
    fig.show()
```

```
import numpy as np
profil_pakai = df.groupby('Nama')['Pakai'].agg(['mean','std']).reset_index()
profil_pakai.columns = ['Nama','hist_mean','hist_std']

df_merged = df.merge(profil_pakai, on='Nama', how='left')
df_merged['z_score'] = ((df_merged['Pakai'] - df_merged['hist_mean'])
                         / df_merged['hist_std'].replace(0, np.nan))

lonjakan = (df_merged[df_merged['z_score'] > 3]
            .sort_values('z_score', ascending=False)
            [['Bulan','Nama','Segmen','Pakai','hist_mean','z_score','Total akhir']])
lonjakan = lonjakan.round(1)

print(f'=== Lonjakan Ekstrim (z-score > 3) ===')
print(f'Jumlah: {len(lonjakan)} kejadian\n')
print(lonjakan.head(20).to_string(index=False))
```
